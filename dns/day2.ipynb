```python
testX = df2['description'][::2]
```

```python
trainX = df2['description'][1::2]
```

```python
testY = df2['category_id'][::2]
```

```python
trainY = df2['category_id'][1::2]
```

```python
# testX
```

```python
cnt_vec = CountVectorizer()
trainX = cnt_vec.fit_transform(trainX)

testX = cnt_vec.transform(testX)
```

```python
# trainX = CountVectorizer().fit_transform(trainX)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
classifier = MultinomialNB().fit(trainX,trainY)
```

```python
classifier
```

```python
print("{0:.4f}".format(classifier.score(trainX, trainY)))
```

```python
print("{0:.4f}".format(classifier.score(testX, testY)))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dec_tree = DecisionTreeClassifier().fit(trainX, trainY)

print("{0:.4f}".format(dec_tree.score(trainX, trainY)))
print("{0:.4f}".format(dec_tree.score(testX, testY)))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
predicted = dec_tree.predict(testX)
```

```python
confusion_matrix(testY, predicted)
```
